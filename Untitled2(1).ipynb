```python
#7 fun
def det_out(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1

    low = q1 - 1.5 * iqr
    up = q3 + 1.5 * iqr

    fil_df = df[(df[col] >= low) & (df[col] <= up)]
    outl = df[(df[col] < low) | (df[col] > up)]

    return fil_df, outl
```
